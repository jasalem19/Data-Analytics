Exercise 1

```python
%%sql

SELECT p.productid,p.name , AVG(pr.rating) AS avgrating ,
COUNT(pr.rating) AS num_ratings
FROM
    product p
INNER JOIN
    productReview pr ON p.productId = pr.productId
GROUP BY
    p.productId
ORDER BY
    avgrating DESC,p.productid ASC;
```

Exercise 2.1

```python
%%sql
SELECT pmpdc.productModelID, pd.description
FROM productModelProductDescriptionCulture AS pmpdc
JOIN productDescription AS pd ON pmpdc.productDescriptionID = pd.productDescriptionID
WHERE pmpdc.cultureID = 'en';
```

Exercise 2.2

```python
%%sql
with english_description as (SELECT pmpdc.productModelID, pd.description
FROM productModelProductDescriptionCulture AS pmpdc
JOIN productDescription AS pd
ON pmpdc.productDescriptionID = pd.productDescriptionID
WHERE pmpdc.cultureID = 'en')

select e.productModelId, e.description, p.name as Name, sum(s.orderqty) as total_orders 
from product as p
join salesOrderDetail as s on s.productId = p.productId
join english_description as e on e.productModelId = p.productModelId
group by p.productId
order by total_orders desc
limit 10;
```

Exercise 3.2

```python
%%sql
select p.productId, c.name as Category, s.name as Subcategory, p.listPrice
from product as p
join productSubcategory as s on s.productSubcategoryId = p.productSubcategoryId
join productCategory as c on c.productCategoryId = s.productCategoryId
group by p.productId;

```

Exercise 3.3

```python
%%sql
with productQuantities as
(select p.productId , sum(s.orderqty) as total_orders 
from product as p
join salesOrderDetail as s on s.productId = p.productId
group by p.productId),

productPrices as (
select p.productId, c.name as Category, s.name as Subcategory, p.listPrice
from product as p
join productSubcategory as s on s.productSubcategoryId = p.productSubcategoryId
join productCategory as c on c.productCategoryId = s.productCategoryId
group by p.productId)

select pp.category, pp.Subcategory, avg(pp.listPrice) average_price_in_subcategory, sum(q.total_orders) total_items_sold_in_subcategory
from productQuantities as q
join productPrices as pp on pp.productId = q.productId
group by pp.Subcategory
order by pp.category;
```

Exercise 6.2

```python
%%sql
with salesOrderDetailsCTE as 
(select salesOrderID,sum(orderQty*(unitPrice - unitPrice*unitPriceDiscount))
as ordertotal
from salesOrderDetail 
group by salesOrderID), 

salesOrderHeaderCTE as 
(select salesPersonID,salesOrderID
from salesOrderHeader
where orderDate >= "2014-01-01" and orderDate <= "2014-12-31" and salesPersonID is not NULL and salesPersonID <> "")

select h.salesPersonID, sum(d.ordertotal) as ordertotalsum
from salesOrderDetailsCTE as d
join salesOrderHeaderCTE as h on h.salesOrderID = d.salesOrderID
group by h.salesPersonID
order by ordertotalsum desc
limit 5;
```

Exercise 7

```python
%%sql

with salesPersonTotalSales as (
with orders as 
(select salesOrderID,sum(orderQty*(unitPrice - unitPrice*unitPriceDiscount))
as ordertotal
from salesOrderDetail 
group by salesOrderID), 

salesPersonAndOrders as 
(select salesPersonID,salesOrderID
from salesOrderHeader
where orderDate >= "2014-01-01" and orderDate <= "2014-12-31" and salesPersonID is not NULL and salesPersonID <> "")

select h.salesPersonID, sum(d.ordertotal) as ordertotalsum
from orders as d
join salesPersonAndOrders as h on h.salesOrderID = d.salesOrderID
group by h.salesPersonID
order by ordertotalsum desc
)

select spts.salesPersonId, spts.orderTotalSum, sp.commissionpct
from salesPersonTotalSales as spts
join salesPerson as sp on sp.businessEntityID = spts.salesPersonID;
```

Exercise 8.1

```python
%%sql
select h.salesPersonID, h.salesOrderID, c.CurrencyRateID, c.toCurrencyCode
from salesOrderHeader as h
left join currencyRate as c on h.currencyRateID = c.currencyRateID
where h.salesPersonID <> "" and salesPersonID is not null and h.orderDate like "%2014%"
order by h.salesPersonID
limit 10;
```

Exercise 8.2

```python
%%sql
select h.salesPersonID, h.salesOrderID,
CASE 
when c.toCurrencyCode is null then 'USD'
else c.toCurrencyCode
end as toCurrencyCode
from salesOrderHeader as h
left join currencyRate as c on h.currencyRateID = c.currencyRateID
where h.salesPersonID <> "" and salesPersonID is not null and h.orderDate like "%2014%"
order by h.salesPersonID
limit 10;
```

Exercise 9

```python
%%sql

with salesPersonTotalSales as (
with orders as 
(select salesOrderID,sum(orderQty*(unitPrice - unitPrice*unitPriceDiscount))
as ordertotal
from salesOrderDetail 
group by salesOrderID), 

salesPersonAndOrders as 
(select h.salesPersonID,h.salesOrderID,
CASE 
when c.toCurrencyCode is null then 'USD'
else c.toCurrencyCode
end as toCurrencyCode
from salesOrderHeader as h
left join currencyRate as c on h.currencyRateID = c.currencyRateID
where h.salesPersonID <> "" and salesPersonID is not null and h.orderDate like "%2014%"
)

select h.salesPersonID, sum(d.ordertotal) as ordertotalsum,h.toCurrencyCode
from orders as d
join salesPersonAndOrders as h on h.salesOrderID = d.salesOrderID
group by h.salesPersonID
)

select spts.salesPersonId, spts.toCurrencyCode, spts.orderTotalSum, sp.commissionpct
from salesPersonTotalSales as spts
join salesPerson as sp on sp.businessEntityID = spts.salesPersonID
order by spts.toCurrencycode ASC,sp.commissionpct DESC;
```
